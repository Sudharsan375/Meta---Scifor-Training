#Question part 1

```python
import pandas as pd
import numpy as np
```

data frame creation

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Score': [85.5, 90.3, 88.0, 92.1, 78.8]
}

df = pd.DataFrame(data)
```

inspecting data

```python
print("first three rows:")
df.head(3)
```

column selection

```python
print("name & score column:")
df[['Name', 'Score']]
```

filtering data

```python
print("rows where age is greater than 25:")
df[df['Age'] > 25]
```

adding new column

```python
df['Passed'] = df['Score'] > 80
print("adding new column:")
print(df)
```

sorting data

```python
print("sorted by score in descending order:")
df.sort_values(by='Score', ascending=False)
```

group by

```python
print("group by mean score for each Passed group:")
df.groupby('Passed')['Score'].mean()
```

handling Missing Data

```python
print("Add a new column Attendance ")
df['Attendance'] = [90, np.nan, 85, 87, np.nan]
df
```

```python
print("missing values with the mean of the column :")
df['Attendance'] = df['Attendance'].fillna(df['Attendance'].mean())
df
```

#Question part 2

array creation

```python
print("Array with values ranging from 10 to 19:")
array = np.arange(10, 20)
print(array)
```

array reshape

```python
print("Reshaped array:")
reshaped_array = array.reshape(2, 5)
print(reshaped_array)
```

array operation

```python
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

print("element-wise addition:")
print(array1 + array2)
print("element-wise multiplication:")
print(array1 * array2)
```

 Array Indexing

```python
print("Second and fourth elements:")
array = np.array([10, 20, 30, 40, 50])
print(array[1], array[3])
```

Array Slicing

```python
array = np.array([1, 2, 3, 4, 5, 6])
print("Sliced array:")
print(array[1:4])
```

Broadcasting

```python
array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

print("Broadcasting example:")
print(array1 + array2)
```

Matrix Operations

```python
identity_matrix = np.eye(3, dtype=int)
print("Identity matrix (3x3):")
print(identity_matrix)
```

#Question part 3

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Creating Box Plot

```python
data = [12, 15, 14, 10, 8, 11, 14, 18, 22, 24, 30, 28, 26, 30, 32, 35, 38]
plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title("Box Plot")
plt.show()
```

Creating Histogram

```python
plt.figure(figsize=(8, 6))
plt.hist(data, bins=10,edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

Creating Bar Plot

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=df.index, y=df['Score'])
plt.title("Bar Plot")
plt.xlabel("Individuals")
plt.ylabel("Score")
plt.xticks(rotation=90)
plt.show()
```

Creating Pie Plot

```python
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [10, 15, 7, 3]
plt.figure(figsize=(8, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Pie Plot")
plt.axis('equal')
plt.show()
```

1) Write a function `fibonacci(n)` that generates the first `n` numbers in the Fibonacci sequence and return

```python
def fibonacci(n):
    if n<=0:
        return []
    elif n==1:
        return [0]
    elif n==2:
        return [0,1]
    else:
      fib = [0, 1]
      for i in range(2, n):
          fib.append(fib[i-1] + fib[i-2])
      return fib

n=int(input("Enter the number : "))
print(fibonacci(n))
```

2. Write a function `is_prime(num)` that checks if a given number is a prime number. Return `True` if the n

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

num=int(input("Enter the number : "))
print(is_prime(num))
```

```python
import pandas as pd

data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Sales': [200, 220, 210, 215, 225],
    'Region': ['North', 'South', 'East', 'West', 'North'],
    'Manager': ['John', 'Jane', 'Joe', 'Jake', 'John']
}


df = pd.DataFrame(data)
print("display the first 5 rows")
print(df.head())

df['Profit'] = df['Sales'] * 0.1
print("\nthe new 'Profit' column:")
print(df)

region_sales = df.groupby('Region')['Sales'].sum().reset_index()
print("\nCalculate total sales for each region")
print(region_sales)

avg_sales = df.groupby('Region')['Sales'].mean().reset_index()
print(avg_sales)
print(f"\nThe region with the highest average sales is {avg_sales.loc[avg_sales['Sales'].idxmax()]['Region']}")
```

```python
import pandas as pd

# 1. Load the CSV file from GitHub
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
```

```python
# Display the first 5 rows
print("First 5 rows:")
print(df.head())
```

```python
# 2. Display summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# 3. Check for null values
print("\nNull value count:")
print(df.isnull().sum())
```

```python
# 4. Fill null values in 'Age' column with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
df
```

```python
# 5. Drop rows with null 'Embarked'
df = df.dropna(subset=['Embarked'])
df
```

```python
# 6. Create new column 'FamilySize'
df['FamilySize'] = df['SibSp'] + df['Parch']
df
```

```python
# 7. Convert 'Sex' column to numerical values
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df
```

```python
# 8. Group by 'Pclass' and calculate average fare
print("\nAverage fare by Pclass:")
df.groupby('Pclass')['Fare'].mean()
```

```python
# 9. Find the most common port of embarkation
print("\nMost common port of embarkation:")
df['Embarked'].mode()[0]
```

```python
# 10. Create a pivot table for survival rate by Sex and Pclass
print("\nSurvival rate by Sex and Pclass:")
pivot_table = df.pivot_table(index='Sex', columns='Pclass', values='Survived', aggfunc='mean')
pivot_table
```
